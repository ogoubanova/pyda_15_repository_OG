```python
import pandas as pd
import numpy as np
```

```python
# Выводы:
#     'age' -- в данных опечатка: вместо '2' стоит значение '9', следует заменить, 
#      тогда ср.значение, медиана и мода == 1. 
#     'rec_temp', 'pulse' - средняя ректальная температура 38.16 градусов, пульс 71.93 медиана=64), 
#     'extr_temp' - ср. температура конечностей: 2 - 'теплые', значение медианы=3 (холодные)
#     'muc_membranes' -- ср. ззначение 2.85 (медиана = 3) - слизистые бледно-розовоые
#     'pain' -- ср. уровень боли, ближе к 1 ('normal'), то есть без болевого синдрома
#     'peristalsis' -- ср. значение 2.9 (медиана = 3)-- показатель гипомотильного состояния
#     'abd_dist' -- ср. значение 2.26 (медиана = 2)-- легкий запор 

horse_data = pd.read_csv('horse_data.csv',na_values=['?', 'NaN', 'NAN', None, np.nan, np.NaN, np.NAN])

selected_categories = [1, 3, 4, 6, 8, 9, 11, 12]
df = horse_data.iloc[:, selected_categories ]
columns = ['age', 'rec_temp', 'pulse', 'extr_temp', 'muc_membranes', 'pain', 'peristalsis', 'abd_dist']
df.columns = columns
description = df.describe()

description
```

```python
# заменим ошибочное значение '9' на '2'
df.age.replace(9, 2, inplace=True)
df.head(10)
```

```python
# посмотрим, сколько лошадей с повышенной температурой или сердцебиением
print(len(df))
# чуть больше половины лошадей с повышенно температурой
print(df[df.rec_temp > 37.8].count())

# и почти все с повышенным пульсом
df[ (df.rec_temp > 37.8) | (df.pulse > 40)].count()
df[ df.pulse > 40 ].count()
```

```python
# Задание 2. Выбросы.
# В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах 
# и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.
# На основании результатов применения функции describe()к данным df можно заключить, 
# что в столбцах категориальных величин 'pulse', 'extr_temp', 'muc_membranes', 'pain', 'peristalsis', 'abd_dist'
# нет выбросов
```

```python
# Анализ выбросов для столбца 'age'
df.age.head()

firstQ = df.age.quantile(0.25)
thirdQ = df.age.quantile(0.75)

IQR = thirdQ - firstQ
low_outlier = firstQ - 1.5*IQR
high_outlier = thirdQ + 1.5*IQR

print(low_outlier)
print(high_outlier)

# данные о молодых лошадях (24 строки) нужно оставить
# превалируют данные о взрослых лошадях: болеют в основном старенькие лошадки
# низких выбросов нет
print(df.age[ df.age > high_outlier].count())
print(df.age[ df.age < low_outlier].count())
```

```python
# Анализ выбросов для столбца 'pulse'

firstQ = df.pulse.quantile(0.25)
thirdQ = df.pulse.quantile(0.75)

IQR = thirdQ - firstQ
low_outlier = firstQ - 1.5*IQR
high_outlier = thirdQ + 1.5*IQR

print(low_outlier)
print(high_outlier)

# 5 выбросов -- исходя из других показателей, можно сделать вывод, что причиной выбросов яляется
# наличие данных об очень больных животных
# поскольку количество выросов мало: 1.7%, то можно их оставить

print(f'Number of high outliers: {df.pulse[ df.pulse > high_outlier].count()}')

# no low outliers
print(f'Number of low outliers: {df.pulse[ df.pulse < low_outlier ].count()}')

# у большинства лошаде повышенный пульс, у 5 очень высокий. Но это естесственные выбросы, 
# связанные с болезнье, следует оставить в данных
print(f'Number of horses with elevated pulse: {df.pulse[ df.pulse > 40 ].count()}')
```

```python
# Анализ выбросов для столбца 'rec_temp' 
# выбросов нет
firstQ = 0.5*df.rec_temp.median()
thirdQ = 1.25*df.rec_temp.median()
IQR = thirdQ - firstQ
low_outlier = firstQ - 1.5*IQR
high_outlier = thirdQ + 1.5*IQR


print(low_outlier)
print(high_outlier)

# no low/high outliers
print(df.rec_temp[ df.rec_temp > high_outlier])
print(df.rec_temp[ df.rec_temp < low_outlier])
```
