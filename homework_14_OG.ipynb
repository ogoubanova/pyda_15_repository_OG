```python
# Задание 14. продвинутый pandas 2
# Задание 1
# Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок.
# Под временем жизни понимается разница между максимальным и минимальным значением 
# столбца timestamp для данного значения userId.
```

```python
import pandas as pd

ratings = pd.read_csv('../ml-latest-small/ratings.csv')
# ratings.head(10)
```

```python
def calc_time_lap(data):
    if data.rating.count() > 100:
        return data.timestamp.max() - data.timestamp.min()
```

```python
obj = ratings.groupby('userId').apply(calc_time_lap)
mean_lap_time = round(obj.mean()/3600/24/365, ndigits=2)
print(f'Среднее время жизни пользователей, которые выставили более 100 оценок: {mean_lap_time} лет')
```
