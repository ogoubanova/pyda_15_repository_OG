```python
## Задание 2
## По данным файла power.csv посчитайте суммарное потребление 
## стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 
## за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

import pandas as pd

data = pd.read_csv('power.csv')
data.head()

def baltic(country):

    if country in ['Lithuania', 'Latvia', 'Estonia']:
        return 'Прибалтика'
    return 'other'

data['baltic'] = data.country.apply(baltic)

blt_countries = data[ (data['baltic'] == 'Прибалтика') & (data['year'] >= 2005) & (data['year'] <= 2008) \
  & ( (data['category'] == 4) | (data['category'] == 12) | (data['category'] == 21) ) ]\
  .sort_values(by=['country', 'year', 'quantity'], ascending=[True, True, False])

# print(f'Size of data: {len(blt_countries)}')

blt_consumption = blt_countries['quantity'].sum()
print(f'Суммарное потребление стран Балтики: {blt_consumption}')

```
