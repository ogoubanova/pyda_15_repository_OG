```python
## Домашнее задание 12. Функции pandas. ОГ
## Задание 1.

import pandas as pd

def rate_movie(rt_str):
    """
    классифицирует фильмы по следующим правилам:
    - оценка 2 и меньше - низкий рейтинг
    - оценка 4 и меньше - средний рейтинг
    - оценка 4.5 и 5 - высокий рейтинг
    """
    
    rt_num = float(rt_str)
    if rt_num <= 2.0:
        return 'низкий рейтинг'
    elif rt_num > 2.0 and rt_num <= 4.0:
        return 'средний рейтинг'
    elif rt_num > 4.5 and rt_num <= 5.0:
        return 'высокий рейтинг'
    else:
        "нет ретинга"

ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings['class'] = ratings.rating.apply(rate_movie)
ratings.head(10)
```
